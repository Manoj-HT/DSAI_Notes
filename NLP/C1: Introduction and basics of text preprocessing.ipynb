# INTRODUCTION AND BASICS OF TEXT PREPROCESSING

## Introduction to Natural Language Processing

- **Definition**: Natural Language Processing (NLP) is a subfield of AI that enables computers to understand, interpret, and generate human language.  
- **Goal**: To make machines interact with humans in natural language.

### Need for NLP
- Computers can understand only structured data.  
- Most human knowledge is unstructured (e.g., emails, books, social media, etc.).  
- Due to this unstructured nature and the inability of machines to understand it, much of this data remains locked.  
- We need NLP to bridge the gap so that machines can start understanding this data.  
- Human language is full of synonyms, context, sarcasm, grammar variations, etc.  
- Simple keyword-based or rule-based methods fail â†’ NLP is needed to handle this complexity.  
- Humans cannot manually read and process billions of documents, messages, or interactions.  
- NLP automates this understanding at scale.  

### Applications of NLP
- **Search Engines** (Google, Bing â†’ query understanding)  
- **Chatbots/Assistants** (Siri, Alexa, GPT)  
- **Sentiment Analysis** (analyzing reviews, tweets)  
- **Machine Translation** (Google Translate)  
- **Information Extraction** (NER, summarization)  
- **Text Generation** (story writing, code completion)  

## Introduction to Text Preprocessing
- **Definition**: Text preprocessing is the initial step in NLP where raw text is cleaned and transformed into a format that machines can understand and work with effectively.  
- **Goal**: To make text uniform for analysis or model training.  

### Challenges with Text Data
1. **Ambiguity**: Words can have multiple meanings  
   - Eg: *bank* â†’ financial institution vs. river bank  
2. **Synonyms and Variations**: Many ways to express the same idea  
   - Eg: "Car", "Automobile", "Vehicle"  
3. **Context and Semantics**: Meaning often depends on surrounding words  
   - Eg: "I saw her duck"  
4. **Noisy Data**: Text often contains slang, typos, abbreviations, emojis, hashtags, etc.  
   - Eg: "gr8 m8 lol ðŸ˜‚"  
5. **Stop Words & Irrelevant Information**: Words like "this", "is", "and" add bulk to the text with little meaning.  
6. **High Dimensionality**: Large vocabulary size leads to sparse representations when converting text to vectors.  
7. **Morphological Variations**: Words change form based on tense.  
   - Eg: "run", "running", "ran"  
8. **Multilinguality**: Mixing words from different languages in one text.  
   - Eg: "Hola bro, como estas"  
9. **Domain-Specific Language**: Specialized terms in certain fields may not be understood by general models.  
   - Eg: medical terms, legal terms, financial terms  
10. **Sarcasm and Figurative Language**: Text containing irony, humor, or metaphors is hard for machines to interpret.  
   - Eg: "Oh great, another Monday"  

### Importance of Text Preprocessing
1. Converts raw text into a usable form  
2. Reduces noise  
3. Handles ambiguity  
4. Reduces variations  
5. Boosts accuracy of NLP models  
6. Enables numerical representation  


```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

text = "NLP is amazing! It allows computers to understand human language"
print("\n","Sentence tokenization", sent_tokenize(text), "\n")
print("\n", "Word Tokenization", word_tokenize(text), "\n")
```

## Common Text Preprocessing / Cleaning Methods

| Method                                   | What it does                               | Example                            |
| ---------------------------------------- | ------------------------------------------ | ---------------------------------- |
| **Lowercasing**                          | Standardizes text                          | "HELLO" â†’ "hello"                  |
| **Tokenization**                         | Splits text into words/sentences           | "I love NLP" â†’ ["I", "love", "NLP"] |
| **Stopword Removal**                     | Removes common but uninformative words     | "the", "is", "a"                   |
| **Stemming**                             | Reduces words to their root form (fast, rough) | "running" â†’ "run"                  |
| **Lemmatization**                        | Converts words to root form using vocabulary & grammar (more accurate than stemming) | "better" â†’ "good" |
| **Removing Punctuation & Special Characters** | Cleans unnecessary symbols                | "hello!!!" â†’ "hello"               |
| **Handling Emojis/Slang**                | Normalizes or removes emojis/slang         | ðŸ™‚ â†’ "smile"                       |
| **Handling Spelling Mistakes**           | Corrects misspelled words                  | "recieve" â†’ "receive"              |
| **Handling Numbers**                     | Removes, normalizes, or replaces numbers   | "I ran 10km" â†’ "I ran NUM"         |
| **Removing Extra Whitespace**            | Normalizes spacing                         | "I   love   NLP" â†’ "I love NLP"    |


```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import re

nltk.download('stopwords')
nltk.download('wordnet')

text = "NLP is an amazing field! It involves processing, understanding, and generating languages."

# Lowercase
text = text.lower()

# Remove punctuation
text = re.sub(r'[^\w\s]', '', text)

# Tokenization
words = nltk.word_tokenize(text)

# Stopword removal
stop_words = set(stopwords.words('english'))
filtered = [w for w in words if w not in stop_words]

# Stemming and Lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("\n", "Original words:", words, "\n")
print("\n", "After stopwprd removal:", filtered ,"\n")
print("\n", "Stemming:", [stemmer.stem(word) for word in filtered])
print("\n", "Lemmatizing:", [lemmatizer.lemmatize(word) for word in filtered])
```

## Introduction to Web Scraping: BeautifulSoup

- **Definition**: Web scraping is the process of collecting text data from websites for NLP tasks.  
- **Steps**:  
  1. Fetch the HTML content using the `requests` library  
  2. Parse the HTML with **BeautifulSoup**  
  3. Extract relevant text (e.g., from `<p>` or `<div>` tags)  


```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Natural_language_processing"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

paragraphs = soup.find_all('p')
for p in paragraphs[:2]:
    print(p.get_text())
```

## Named Entity Recognition (NER)

- **Definition**: Named Entity Recognition is the process of identifying and classifying named entities (such as people, places, organizations, dates, etc.) in a given text.  
- **Example**:  
  - Input: "Barack Obama was born in Hawaii in 1961"  
  - Output:  
    - Person: Barack Obama  
    - Location: Hawaii  
    - Date: 1961  


```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Barack Obama was born in Hawaii in 1961")

for entity in doc.ents:
    print(entity.label_, ": ", entity.text)
```

## Word Cloud

- **Definition**: A word cloud is a visual representation of text data where the size of each word indicates its frequency or importance in the text.  
- **Use**: Helps quickly identify the most frequent or significant words in a dataset.  


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Natural Language Processing makes human language understandable for computers. NLP is widely used in chatbots, translation, and sentiment analysis."

wc = WordCloud(width=600, height=400, background_color="white").generate(text)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
