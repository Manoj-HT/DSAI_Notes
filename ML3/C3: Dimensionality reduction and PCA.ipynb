# C3: Dimensionality reduction and PCA

## Dimensionality
- **Definition:** The number of features in a dataset  
- **Example:** A dataset with `age`, `height`, and `weight`

### Types
- **Low dimensional:** Few features  
- **High dimensional:** Many features

### Curse of Dimensionality
In higher dimensions:  
- Data points appear equally far apart  
- It becomes harder to find meaningful clusters or neighbors  
- Computation cost increases

### Dimensionality Reduction
- **Key Idea:** Reduce the number of features while keeping most of the information  
- Two approaches:  
  - **Feature Selection:** Pick the most important features  
  - **Feature Extraction:** Create new features from existing ones

### Common Dimensionality Reduction Techniques
- Principal Component Analysis (PCA)  
- t-SNE / UMAP  
- Feature Selection  

## Principal Component Analysis (PCA)
- A feature extraction method  
- Finds the directions where data varies the most  

### Working
1. Standardize the data  
2. Compute the covariance matrix (measures how features vary together)  
3. Find eigenvalues and eigenvectors  
4. Sort by importance and keep top components with the most variance  
5. Transform data into a smaller feature space  

### Advantages
1. Reduces dimensionality  
2. Removes noise and redundancy  
3. Helps with visualization  
4. Prevents overfitting by ignoring less important variance  

## Covariance Matrix
- **Definition:** A square matrix showing how different features vary together  
- Each entry represents the relationship between two features  

### Structure
For $n$ features, the covariance matrix is $n \times n$:  

$\mathrm{Cov(X)} = \begin{bmatrix} 
Var(x_1) & Cov(x_1,x_2) & \dots & Cov(x_1, x_n) \\
Cov(x_2,x_1) & Var(x_2) & \dots & Cov(x_2, x_n) \\ 
\dots & \dots & \dots & \dots \\
Cov(x_n,x_1) & Cov(x_n,x_2) & \dots & Var(x_n) \\ 
\end{bmatrix}$  

- **Diagonal elements:** Variance of each feature  
- **Off-diagonal elements:** Covariance between feature pairs  

### Relationships
- **Positive covariance:** Features increase together  
- **Negative covariance:** One feature increases while the other decreases  
- **Zero covariance:** No linear relationship  

### Importance in PCA
- PCA uses the covariance matrix to find principal components  
- **Eigenvectors:** Represent new feature directions  
- **Eigenvalues:** Indicate the importance of each direction  


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Apply PCA: reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Explained variance ratio:", pca.explained_variance_ratio_)

# Plot PCA results
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=40)
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA on Iris Dataset")
plt.colorbar(label="Species")
plt.show()
```
