# C4: PCA - CASE STUDY

## Why PCA works

- In real world datasets, not all features are equally useful
- Some features carry signal - important variations
- Others are mostly noise - random fluctuations

PCA keeps the directions (principal components) where signal(variance) is highest and ignores low variance directions (likely noise)

## Eigenvectors and Eigenvalues

PCA uses linear algebra
- The covariance matrix tells how features vary together
- Eigenvectors show the direction of max variance
- Eigen values tell how much cariance os explained along each direction

## Case study

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

# Load digits dataset (1797 samples, 64 features = 8x8 images)
digits = load_digits()
X = digits.data
y = digits.target

# Apply PCA: reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Explained variance ratio:", pca.explained_variance_ratio_.sum())

# Plot PCA results
plt.figure(figsize=(10,8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=20)
plt.legend(*scatter.legend_elements(), title="Digits")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA on Handwritten Digits Dataset (64D  2D)")
plt.show()
```

## When PCA is useful

- **Visualization :** When high dimensional data is present go for 3d or 3d plots
- **Speeding up models :** Fewer features helps train model faster
- **Noise reduction :** Drop low variance components to reduce noise
- **Preprocessing :** This is done before clustering usefull for recommendation systems
- **Limitations :** PCA can be only found on linear data, it struggles on non linear data
